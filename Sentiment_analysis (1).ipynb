```python
df = pd.read_csv('social_media.csv')  # use actual filename
df.head()
```

```python
from textblob import TextBlob

# Calculate polarity for each text
def get_polarity(text):
    return TextBlob(str(text)).sentiment.polarity

df['Polarity'] = df['Text'].apply(get_polarity)
```

```python
def get_sentiment(polarity):
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['Predicted_Sentiment'] = df['Polarity'].apply(get_sentiment)
```

```python
import matplotlib.pyplot as plt

df['Predicted_Sentiment'].value_counts().plot(kind='pie', autopct='%1.1f%%',
                                               colors=['lightblue', 'lightgreen', 'salmon'])
plt.title("Predicted Sentiment Distribution")
plt.ylabel("")
plt.show()
```

```python
import seaborn as sns

sns.countplot(data=df, x='Predicted_Sentiment', palette='pastel')
plt.title("Number of Posts by Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```
